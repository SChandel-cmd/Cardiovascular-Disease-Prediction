```python
import pickle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, plot_roc_curve
import matplotlib.pyplot as plt
import math
```

```python
path="/content/gdrive/MyDrive/DE Project/"
```

```python
X_train4 = pickle.load(open(path + 'X train4.pkl','rb'))
y_train4 = pickle.load(open(path+ 'y train4.pkl','rb'))
X_test4 = pickle.load(open(path + 'X test4.pkl','rb'))
y_test4 = pickle.load(open(path+ 'y test4.pkl','rb'))
```

```python

X_train1 = pickle.load(open(path + 'X train1.pkl','rb'))
y_train1 = pickle.load(open(path+ 'y train1.pkl','rb'))
X_test1 = pickle.load(open(path + 'X test1.pkl','rb'))
y_test1 = pickle.load(open(path+ 'y test1.pkl','rb'))

X_train2 = pickle.load(open(path + 'X train2.pkl','rb'))
y_train2 = pickle.load(open(path+ 'y train2.pkl','rb'))
X_test2 = pickle.load(open(path + 'X test2.pkl','rb'))
y_test2 = pickle.load(open(path+ 'y test2.pkl','rb'))

X_train3 = pickle.load(open(path + 'X train3.pkl','rb'))
y_train3 = pickle.load(open(path+ 'y train3.pkl','rb'))
X_test3 = pickle.load(open(path + 'X test3.pkl','rb'))
y_test3 = pickle.load(open(path+ 'y test3.pkl','rb'))
```

```python
model4 = KNeighborsClassifier(n_neighbors=265)
model1 = KNeighborsClassifier(n_neighbors=265)
model2 = KNeighborsClassifier(n_neighbors=265)
model3 = KNeighborsClassifier(n_neighbors=265)
```

```python
model4.fit(X_train4, y_train4)
predictions4 = model4.predict(X_test4)
```

```python
model1.fit(X_train1, y_train1)
predictions1 = model1.predict(X_test1)
```

```python
model2.fit(X_train2, y_train2)
predictions2 = model2.predict(X_test2)
```

```python
model3.fit(X_train3, y_train3)
predictions3 = model3.predict(X_test3)
```

```python
accuracy4 = accuracy_score(y_test4,predictions4)
CM4 = confusion_matrix(y_test4,predictions4)
TN4 = CM4[0][0]
FN4 = CM4[1][0]
TP4 = CM4[1][1]
FP4 = CM4[0][1]
N4 = FP4+TN4
P4 = TP4+FN4
sens4 = TP4/(P4)
spec4 = TN4/N4
fscore4 = TP4/(TP4+((FP4+FN4)/2))

FDR4 = FP4/(TP4+FP4)
FNR4 = FN4/P4
NPV4 = TN4/(TN4+FN4)
FPR4 = FP4/N4
prec4 = TP4/(TP4+FP4)
MCC4 = ((TP4*TN4)-(FP4*FN4))/math.sqrt((TP4+FP4)*(P4)*(TN4+FN4)*(N4))
```

```python
accuracy1 = accuracy_score(y_test1,predictions1)
CM1 = confusion_matrix(y_test1,predictions1)
TN1 = CM1[0][0]
FN1 = CM1[1][0]
TP1 = CM1[1][1]
FP1 = CM1[0][1]
N1 = FP1+TN1
P1 = TP1+FN1
sens1 = TP1/(P1)
spec1 = TN1/N1
fscore1 = TP1/(TP1+((FP1+FN1)/2))

FDR1 = FP1/(TP1+FP1)
FNR1 = FN1/P1
NPV1 = TN1/(TN1+FN1)
FPR1 = FP1/N1
prec1 = TP1/(TP1+FP1)
MCC1 = ((TP1*TN1)-(FP1*FN1))/math.sqrt((TP1+FP1)*(P1)*(TN1+FN1)*(N1))
```

```python
splits = ['60', '70', '80']
```
